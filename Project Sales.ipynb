```python
import pandas as pd

# Opening the Excel File Dataset Sales
sales = pd.read_excel('E:\Progamming Project\Portofolio\Project_Sales\Sales.xlsx')
print(sales.shape)
sales.set_index('Row ID', inplace=True)
sales.head()
```

```python
# Checking column data types and empty/filled values
sales.info()
```

```python
# Finding the number of empty values in each column
sales.isnull().sum()
```

```python
# Filling empty data with 'Empty' in the 'Postal Code' column
sales['Postal Code'].fillna('Empty', inplace=True)
sales.loc[sales['Postal Code'] == 'Empty', ['Postal Code']]
```

```python
# Checking the data structure again to ensure there are no empty values
sales.info()
```

```python
# Viewing descriptive statistical summary of the DataFrame or Series
sales.describe()
```

1. What product Category & Subcategory are frequently ordered?

```python
# Grouping by Category while aggregating Frequency Data
category = sales.groupby(
        ['Category']).agg(
            {'Category':['count']})
        
category.columns = category.columns.droplevel(0)
category = category.rename(columns={'count':'Fr_Category'})

category
```

```python
# Grouping by Sub-Category while aggregating Frequency Data for a more specific view
subCategory = sales.groupby(
        ['Category','Sub-Category']).agg(
            {'Sub-Category':['count']})
        
subCategory.columns = subCategory.columns.droplevel(0)
subCategory = subCategory.rename(columns={'count':'Fr_Sub-Category'})

subCategory
```

2. Which products are frequently ordered based on the Segment (User) and product grouping, including their total sales?

```python
# Grouping by Segment, Category, and Sub-Category while aggregating Frequency and Total Sales
Segment = sales.groupby(
    ['Segment','Category','Sub-Category']).agg(
        {'Category':['count'],'Sales':['sum']}
    )
Segment.columns = Segment.columns.droplevel(0)

Segment = Segment.rename(columns={'count':'Frequency','sum':'Total Sales'})
Segment
```

3. What is the difference in each Order-to-Delivery time based on the Shipping Method and Product Category?

```python
# Viewing the Time Difference between Order and Delivery based on Shipping Method & Product Category
sales['Times'] = sales['Ship Date'] - sales['Order Date']
sales['Times'] = sales['Times'].astype('str')
time = sales.groupby(
    ['Category', 'Ship Mode','Times']
    ).agg(
        {'Times':['count']}
)
time.columns = time.columns.droplevel(0)   
time = time.rename(columns={'count':'Frequency'})

# Selecting the Range of Time Differences with a Frequency above 100
time = time.loc[time['Frequency'] >= 100]
time
```

4. In which city are deliveries most frequently made based on the province/state?

```python
# Analyzing frequently occurring delivery locations
city = sales.groupby(
    ['City']).agg(
        {'State':['count']})
city.columns = ['count']

# Summing up the Frequency of States grouped by City
city = city.groupby(['City']).agg({'count': 'sum'})
city.columns = ['sum']
city = city.rename(columns={'sum':'Fr_City'})

# Selecting the Range of delivery Frequencies above 80
# Prioritizing larger Frequencies for more valid and accurate data within a data range
city = city.loc[city['Fr_City'] >= 80]
city
```

5. Who are the customers who frequently purchase items, including their city and region?

```python
# Analyzing customers who frequently purchase products
customer = sales.groupby(['Region', 'Customer Name']).agg({'Customer Name': ['count']})
customer.columns = customer.columns.droplevel(0)
customer = customer.rename(columns={'count': 'Fr_Customer'})

# Selecting data with Frequency greater than or equal to 15 for each purchase
# Prioritizing larger Frequencies for more valid and accurate data within a data range
customer = customer.loc[(customer.Fr_Customer >= 15)]
customer
```

6. How is the revenue based on product category (min, mean, max, sum)?

```python
# Viewing the Comprehensive Sales Calculation Based on Product Category
Category_Sales = sales.groupby(['Category']).agg({'Sales':['min','mean','max','sum']})
Category_Sales.columns = ['min','mean','max','sum']
Category_Sales = Category_Sales.rename(columns={
    'min':'Minimum',
    'mean':'Mean',
    'max':'Maximum',
    'sum':'Total'
    })
Category_Sales
```

7. How is the revenue based on product subcategories (min, mean, max, sum)?

```python
# Performing Sales Calculation Based on Product Subcategories
Sub_Sales = sales.groupby(['Sub-Category']).agg({'Sales':['min','mean','max','sum']})
Sub_Sales.columns = ['min','mean','max','sum']
Sub_Sales = Sub_Sales.rename(columns={
    'min':'Minimum',
    'mean':'Mean',
    'max':'Maximum',
    'sum':'Total'
    })
Sub_Sales
```

8. How is the revenue each year (min, mean, max, sum)?

```python
# Extracting Year & Month from the 'Order Date' column into new columns
sales['Year'] = sales['Order Date'].dt.year
sales['Month'] = sales['Order Date'].dt.month

# Performing Sales Calculation Based on Year and Month of Sales
profitYear = sales.groupby(['Year','Month']).aggregate({'Sales':['min','mean','max','sum']})
profitYear.columns = profitYear.columns.droplevel(0)
profitYear = profitYear.rename(columns={
    'min':'Minimum',
    'mean':'Mean',
    'max':'Maximum',
    'sum':'Total'
    })

profitYear
```

9. What is the likelihood of the occurrence of a variation in issues based on Category & Sub-Category (using standard deviation)?

```python
# Finding the Standard Deviation
std = sales.groupby(['Category','Sub-Category']).aggregate({'Sales':['std']})
std.columns = std.columns.droplevel(0)
std = std.rename(columns={'std':'Standard Deviation Sales'})

std
```

10. How many customers choose the shipping method based on segment and product type?

```python
# Grouping by Shipping Method and Aggregating Customer Frequency and Total Sales
method = sales.groupby(['Ship Mode','Segment','Category']).aggregate(
    {'Customer Name':['count'],
     'Sales':['sum']}
    )
method.columns = ['count','sum']
method = method.rename(columns={'count':'Fr_Customer','sum':'Total Sales'})

method
```

```python
# Save the data to an Excel file with each sheet corresponding to the grouping
with pd.ExcelWriter('Project_Sales.xlsx') as writer:
    category.to_excel(writer, sheet_name='Category')
    subCategory.to_excel(writer, sheet_name='Sub-Category')
    Segment.to_excel(writer, sheet_name='Segment')
    time.to_excel(writer, sheet_name='Time')
    city.to_excel(writer, sheet_name='City')
    customer.to_excel(writer, sheet_name='Customer')
    Category_Sales.to_excel(writer, sheet_name='Category_Sales')
    Sub_Sales.to_excel(writer, sheet_name='Sub_Sales')
    profitYear.to_excel(writer, sheet_name='Profit_Year')
    std.to_excel(writer, sheet_name='Standard_Deviation')
    method.to_excel(writer, sheet_name='Method')
```
